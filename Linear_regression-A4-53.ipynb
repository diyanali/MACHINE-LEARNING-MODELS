STEP 4: Plotting the data

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='red',marker='o')
```

STEP 5: Create a linear regression object

```python
reg=linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

STEP 6: Prediction

```python
reg.predict([[5000]]) # Pass the value 3300 as a 2D array with shape (1, 1)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python

m=reg.coef_
b=reg.intercept_
x=3300
y=m*x+b
print(y)
```

```python
d=pd.read_csv('areas.csv')
d
```

```python
p=reg.predict(d)
```

```python
d['prices']=p
```

```python
d.to_csv('prediction1.csv')
```

STEP 6 : Plotting the prediction

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='red',marker='o')
plt.plot(d.area,d.prices,color='blue')
```
